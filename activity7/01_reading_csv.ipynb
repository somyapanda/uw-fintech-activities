```python
import pandas as pd
from pathlib import Path
```

### Read a CSV file as a Pandas DataFrame

Create a path to the file.

```python
csv_path = Path("./Resources/sales.csv")
```

Read the CSV into a DataFrame using Pandas.

```python
sales_df = pd.read_csv(csv_path)
sales_df.head()
```

Read a CSV with no Header.

Without a header in the CSV, Pandas will use the first row as the header.

```python
csv_path = Path("./Resources/sales_no_header.csv")
sales_no_header_df = pd.read_csv(csv_path)
sales_no_header_df.head()
```

Read the file without a header.

```python
sales_df = pd.read_csv(csv_path, header=None)
sales_df.head()
```

Rewrite the column names.

```python
column_names = ["FullName", "Email", "Address", "Zip", "CreditCard", "SalePrice"]
sales_df.columns = column_names
sales_df.head()
```
