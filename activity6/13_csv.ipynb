```python
# Import the necessary libraries for reading csv files
from pathlib import Path
import csv
```

```python
# Set the path for the csv file
csv_path = Path('./Resources/pokemon.csv')
```

```python
# Create new lists to store data for heaviest and tallest pokemon
pokemon_heaviest = []
pokemon_tallest = []
```

```python
# Open the csv
with open(csv_path, 'r') as csvfile:
    # Iterate through the data and search for the number the user inputted. Remember to skip 
    # the header of the CSV.
    csvreader = csv.reader(csvfile, delimiter=',')
    next(csvreader, None)
    for row in csvreader:
        # Print the name of the pokemon(identifier) and pokedex number(species id) at that number. 
        # For example, "Pokemon No. 25 - pikachu".
        identifier = row[1]
        species_id = row[2]
        print(f"Pokemon No.{species_id} - {identifier}")
        # Iterate through the data and search for pokemon whose weight is greater than 3000.
        # Append only the pokemon's name and weight to the 'heaviest' list.
        weight = int(row[4])
        if weight > 3000:
             pokemon_heaviest.append((identifier, weight))
        # Iterate through the data and search for pokemon whose height is greater than 50. 
        # Append only the pokemon's name and height to the 'tallest' list.
        height = int(row[3])
        if height > 50:
             pokemon_tallest.append((identifier, height))
```

```python
print(f"The list of heaviest pokemon is {pokemon_heaviest}.") 
```

```python
print(f"The list of tallest pokemon is {pokemon_tallest}.")                             
```
