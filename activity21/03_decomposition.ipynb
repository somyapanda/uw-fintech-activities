```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Data source: http://course1.winona.edu/bdeppa/FIN%20335/Datasets/US%20Liquor%20Sales.csv

df = pd.read_csv(
    Path("Resources/liquor_sales.csv"), parse_dates=True, index_col="datetime"
)
df
```

```python
df.liquor_sales.plot()
```

```python
df.liquor_sales.loc['1980': '1981'].plot()
```

```python
decomposed = seasonal_decompose(df['liquor_sales'], model='multiplicative')
decomposed.plot()
```

```python
decomposed = seasonal_decompose(df['liquor_sales'], model='additive')
decomposed.plot()
```
