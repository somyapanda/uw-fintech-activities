```python
# Initial imports
import os
import pandas as pd
from newsapi import NewsApiClient
from dotenv import load_dotenv
load_dotenv()
```

```python
# Reading the News API key enviroment variable
api_key = os.getenv("news_api")
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch top news articles about oil
covid_headlines = newsapi.get_top_headlines(q="covid", language="en", country="us")
covid_headlines
```

```python
# Print total articles
print(f"Total articles about covid: {covid_headlines['totalResults']}")

# Show sample article
covid_headlines["articles"][0]
```

```python
# Transform the response dictionary to a DataFrame
covid_df = pd.DataFrame.from_dict(covid_headlines['articles'])
covid_df.head()
```

```python
# Fetch all the news about Facebook Libra
libra_headlines = newsapi.get_everything(q='facebook libra', language='en', sort_by='relevancy')
# Print total articles
print(libra_headlines['totalResults'])
# Show sample article
libra_headlines['articles'][0]
```

```python
# Transform the response dictionary to a DataFrame
libra_df = pd.DataFrame.from_dict(libra_headlines['articles'])
libra_df.head()
```
